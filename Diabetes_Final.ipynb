```python
Gender: This likely indicates the gender of the individual, usually categorized as male or female.

Age: This represents the age of the individual, typically measured in years.

Hypertension: This binary feature likely indicates whether the individual has hypertension (high blood pressure) or not (1 for presence, 0 for absence).

Heart Disease: Similar to hypertension, this binary feature likely indicates whether the individual has heart disease or not.

Smoking History: This feature likely represents the individual's smoking habits, possibly categorized as smoker/non-smoker or with more detail.

BMI (Body Mass Index): BMI is a measure of body fat based on height and weight. It's calculated by dividing a person's weight in kilograms by the square of their height in meters.

HbA1c Level: HbA1c (hemoglobin A1c) is a measure of long-term blood sugar control. It represents the average blood sugar levels over the past 2-3 months.

Blood Glucose Level: This likely represents the individual's current blood glucose level, which is a measure of the amount of glucose in the blood.

Diabetes: This is the target variable and likely indicates whether the individual has diabetes or not (1 for presence, 0 for absence).
```

```python
y=df1['diabetes']
```

```python
del df1['diabetes']
```

```python
x=df1
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy=0.6,random_state=42)
x, y = ros.fit_resample(x, y)
```

```python
y.value_counts()
```

```python
numerical_columns = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x[numerical_columns])
x[numerical_columns] = scaler.transform(x[numerical_columns])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report,f1_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
knn_model=KNeighborsClassifier(n_neighbors=9)
knn_model.fit(x_train,y_train)
y_test_pred=knn_model.predict(x_test)
print("confusion matrix")
print(confusion_matrix(y_test,y_test_pred))
print("accuracy")
print(accuracy_score(y_test,y_test_pred))
print("precision:")
print(precision_score(y_test,y_test_pred))
print("recall")
print(recall_score(y_test,y_test_pred))
print("classification report")
print(classification_report(y_test,y_test_pred))
print("f1_score")
print(f1_score(y_test,y_test_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report,f1_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
logistic_model=LogisticRegression()
logistic_model.fit(x_train,y_train)
y_test_pred=logistic_model.predict(x_test)
print("confusion matrix")
print(confusion_matrix(y_test,y_test_pred))
print("accuracy")
print(accuracy_score(y_test,y_test_pred))
print("precision:")
print(precision_score(y_test,y_test_pred))
print("recall")
print(recall_score(y_test,y_test_pred))
print("classification report")
print(classification_report(y_test,y_test_pred))
print("f1_score")
print(f1_score(y_test,y_test_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report,f1_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
gradient_boosting_model=GradientBoostingClassifier(learning_rate=0.1)
gradient_boosting_model.fit(x_train,y_train)
y_test_pred=gradient_boosting_model.predict(x_test)
print("confusion matrix:")
print(confusion_matrix(y_test,y_test_pred))
print("accuracy:")
print(accuracy_score(y_test,y_test_pred))
print("precision:")
print(precision_score(y_test,y_test_pred))
print("recall:")
print(recall_score(y_test,y_test_pred))
print("classification report")
print(classification_report(y_test,y_test_pred))
print("f1_score:")
print(f1_score(y_test,y_test_pred))
```
